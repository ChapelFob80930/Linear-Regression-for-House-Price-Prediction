```python
#code for splitiing for binary encoder(accuracy=22%(highest as it keeps changing))
from sklearn.model_selection import train_test_split
label_columns=['actualprice']
training_columns=['auto_des_0', 'auto_des_1', 'auto_des_2', 'auto_des_3',
       'auto_des_4', 'auto_des_5', 'auto_des_6', 'bedrooms', 'booking_amt',
       'covAreaUnit', 'flooring_typ_0', 'flooring_typ_1', 'flooring_typ_2',
       'flooring_typ_3', 'flooring_typ_4', 'fur_unfur_0', 'fur_unfur_1',
       'house_facing_0', 'house_facing_1', 'landmarks_0', 'landmarks_1',
       'landmarks_2', 'landmarks_3', 'landmarks_4', 'locality_0', 'locality_1',
       'locality_2', 'locality_3', 'locality_4', 'locality_5',
       'main_charges_0', 'main_charges_1', 'main_charges_2', 'no_balconies',
       'parking_0', 'parking_1', 'parking_2', 'parking_3', 'property_typ_0',
       'property_typ_1', 'seller_type_0', 'seller_type_1', 'seller_type_2',
       'sqftprice']
x_train,x_test,y_train,y_test=train_test_split(df4[training_columns],df4[label_columns],train_size=0.7)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print(x_train.head().to_string())
print(y_train.head().to_string())
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression(fit_intercept = False)
lin_reg.fit(x_train, y_train)

# Score using the test data.
score = lin_reg.score(x_test, y_test)

print('Variance score on test set: {:.0f}%'.format(score * 100))
```
